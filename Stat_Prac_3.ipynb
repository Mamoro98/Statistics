```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('/content/Albw1.csv')
```

```python
df.head()
```

```python
sns.set_theme()
```

```python
import matplotlib.pyplot as plt

sex_counts = df['sex'].value_counts()
sex_counts.index = ['Male','Female']

plt.figure(figsize=(8, 6))
sns.barplot(x=sex_counts.index, y=sex_counts.values,hue=sex_counts)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Distribution of Sex")
plt.show()
```

```python
sns.histplot(data=df,x='LeftEduc16')
```

```python
plt.figure(figsize=(8, 5))
df.boxplot(column='iqfull', by='LeftEduc16', grid=True)
plt.xticks(ticks=[1, 2, 3], labels=['Left After 16', 'Left at/before 16', 'Missing Info'])
plt.ylabel('Full IQ')
plt.title("Full IQ by Mother's Education Level")
plt.suptitle('')
plt.show()
```

```python
before_16 = df['LeftEduc16'] == 0
```

```python
after_16 = df['LeftEduc16'] == 1
```

```python
missing = df['LeftEduc16'] == 2
```

```python
sns.histplot(data=df[before_16]['iqfull'], label="Left at/before 16")
sns.histplot(data=df[after_16]['iqfull'], label="Left After 16")
sns.histplot(data=df[missing]['iqfull'], label="Missing Info")
plt.legend()
plt.show()
```

```python
sns.histplot(data=df[after_16]['iqfull'])
```

```python
sns.violinplot(data=df,x='LeftEduc16',y='iqfull')
plt.xticks(ticks=[0, 1, 2], labels=['Left before 16', 'Left After 16', 'Missing Info'])
plt.xlabel("Mother's Education Level")
plt.ylabel("Full IQ")
plt.title("Full IQ by Mother's Education Level")
plt.show()
```
